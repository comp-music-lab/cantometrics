```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from pathlib import Path
import statsmodels.api as sm
```

```python
df = pd.read_csv('./data/canto-concensus-lomax.csv')
```

```python
df['1'] = round(df['1']/12,2)
df['2'] = round(df['2']/6,2)
df['3'] = round(df['3']/9,2)
df['4'] = round(df['4']/4,2)
df['5'] = round(df['5']/5,2)
df['6'] = round(df['6']/6,2)
df['7'] = round(df['7']/5,2)
df['8'] = round(df['8']/9,2)
df['9'] = round(df['9']/6,2)
df['10'] = round(df['10']/5,2)
df['11'] = round(df['11']/6,2)
df['12'] = round(df['12']/7,2)
df['13'] = round(df['13']/6,2)
df['14'] = round(df['14']/7,2)
df['15'] = round(df['15']/4,2)
df['16'] = round(df['16']/13,2)
df['17'] = round(df['17']/5,2)
df['18'] = round(df['18']/8,2)
df['19'] = round(df['19']/3,2)
df['20'] = round(df['20']/3,2)
df['21'] = round(df['21']/5,2)
df['22'] = round(df['22']/6,2)
df['23'] = round(df['23']/5,2)
df['24'] = round(df['24']/6,2)
df['25'] = round(df['25']/5,2)
df['26'] = round(df['26']/4,2)
df['27'] = round(df['27']/5,2)
df['28'] = round(df['28']/4,2)
df['29'] = round(df['29']/3,2)
df['30'] = round(df['30']/3,2)
df['31'] = round(df['31']/3,2)
df['32'] = round(df['32']/5,2)
df['33'] = round(df['33']/4,2)
df['34'] = round(df['34']/5,2)
df['35'] = round(df['35']/5,2)
df['36'] = round(df['36']/5,2)
df['37'] = round(df['37']/5,2)
```

```python
df= df.set_index('track')
df.index_name = None
pd.DataFrame.to_csv(df, './data/canto-concensus-lomax-normalized.csv')
```

```python
from scipy.spatial.distance import squareform
from scipy.spatial.distance import pdist

df_inv = pd.DataFrame(np.linalg.pinv(df.values), df.columns, df.index)

pairwise = pd.DataFrame(
    squareform(pdist(df, 'euclidean')),
    columns = df.index,
    index = df.index
)
```

```python
pd.DataFrame.to_csv(pairwise, './data/input_matrix.csv')
```

```python
average_triplet_matrix = np.zeros((30, 30))
for i in range(6):
    count = 0
    for j in range(14):
        filename = './data/triple_distance_' + str(i) + chr(65 + j) + '.csv'
        if not Path(filename).is_file():
            continue
        count += 1
        data = np.nan_to_num(pd.read_csv(filename, header=None).to_numpy(), 0)
        average_triplet_matrix[i*5:(i+1)*5, i*5:(i+1)*5] += data
    average_triplet_matrix[i*5:(i+1)*5, i*5:(i+1)*5] /= count
```

```python
triplet = pd.DataFrame(average_triplet_matrix)
```

```python
triplets = {}
inputs = {}
for i in range(6):
    triplets[i] = average_triplet_matrix[i*5:(i+1)*5, i*5:(i+1)*5]
    inputs[i] = pairwise[i*5:(i+1)*5, i*5:(i+1)*5]
```

```python
for value in inputs:
    inputs[value] = inputs[value]*10
```

```python
for i in range(6):
    target_vector = triplets[i].flatten().reshape(-1,1)
    input_vector = inputs[i].flatten().reshape(-1,1)
    X = input_vector
    Y = target_vector
    regr = LinearRegression()
    regr.fit(X, Y)
    model = sm.OLS(Y, X).fit()
    
    print('GROUP %i'%i)
    print('Intercept: \n', round(regr.intercept_[0],4))
    print('Coefficients: \n', round(regr.coef_[0][0],4))
    print(model.summary())
    print("=======================================================================================\n"*3)
```

```python
average_evaluation_matrix = np.zeros((30, 30))
for i in range(6):
    count = 0
    for j in range(14):
        filename = './data/evaluation_' + str(i) + chr(65 + j) + '.csv'
        if not Path(filename).is_file():
            continue
        count += 1
        
```
